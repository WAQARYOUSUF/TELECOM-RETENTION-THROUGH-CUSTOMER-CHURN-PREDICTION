# MODEL BUILDING


```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("tel_churn.csv")
df.head()
```

```python
df=df.drop('Unnamed: 0',axis=1)
```

```python
x=df.drop('Churn',axis=1)
x
```

```python
y=df['Churn']
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# Model 1 DECISION TREE 

```python
model_dt=DecisionTreeClassifier(criterion = "gini",random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model_dt.fit(x_train,y_train)
```

```python
y_pred=model_dt.predict(x_test)
y_pred
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

# HYPERPARAMETER OPTIMIZATION #

```python

from sklearn.model_selection import GridSearchCV


# Define hyperparameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [4, 6, 8, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 4, 8], 
    'criterion': ['gini', 'entropy']  
}

rf_HPO_model = RandomForestClassifier(random_state=100)

grid_search = GridSearchCV(rf_HPO_model, param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=2)
grid_search.fit(x_train, y_train)

best_rf_model = grid_search.best_estimator_

# Predict on test data
y_pred = best_rf_model.predict(x_test)

# Compute accuracy
best_rf_model.score(x_test,y_test)
print(f'Best Random Forest Parameters: {grid_search.best_params_}')

```

 since the best parameters for the random forest classifier are {'criterion': 'entropy', 'max_depth': 8, 'min_samples_leaf': 1, 'min_samples_split': 5, 'n_estimators': 100} we'll train our random forest again with these specific parameters in the below moinority sampling 

```python
print(best_rf_model.score(x_test,y_test))
```

# PRINCIPAL COMPONENT ANALYSIS

```python
from sklearn.decomposition import PCA
pca = PCA(0.9)
xr_train_pca = pca.fit_transform(xr_train1)
xr_test_pca = pca.transform(xr_test1)
explained_variance = pca.explained_variance_ratio_
```

```python
model=RandomForestClassifier(n_estimators=100, criterion='gini', random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model.fit(xr_train_pca,yr_train1)
```

```python
yr_predict_pca = model.predict(xr_test_pca)
```

```python
model_score_r_pca = model.score(xr_test_pca, yr_test1)
```

```python
print(model_score_r_pca)
print(metrics.classification_report(yr_test1, yr_predict_pca))
```

```python
print(f"Number of components selected: {pca.n_components_}")

# 2. See the explained variance ratio per component
print("Explained variance ratio per component:")
print(pca.explained_variance_ratio_)

# 3. See which original features contribute most to each PC
# Create a DataFrame showing feature loadings
loadings = pd.DataFrame(
    pca.components_.T,  # Transpose to get features as rows
    columns=[f'PC{i+1}' for i in range(pca.n_components_)],
    index=X_resampled1.columns  # Original feature names
)

# Show top 5 features contributing to each principal component
print("\nTop features per principal component:")
for pc in loadings.columns:
    print(f"\n{pc}:")
    print(loadings[pc].abs().sort_values(ascending=False).head(5))
```
